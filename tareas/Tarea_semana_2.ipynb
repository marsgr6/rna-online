# üß† Tarea Semana 2: Regresi√≥n vs Clasificaci√≥n con Redes Neuronales (Low Code)

Este notebook est√° dise√±ado para resolver un problema supervisado (clasificaci√≥n o regresi√≥n) utilizando un modelo de red neuronal (`MLPClassifier` o `MLPRegressor`), y aplicar optimizaci√≥n de hiperpar√°metros con validaci√≥n cruzada (10 folds).

‚û°Ô∏è **Recuerda cambiar `var_y` por el nombre real de tu variable objetivo.**


```python
# üì• Cargar datos
import pandas as pd
df = pd.read_csv('tus_datos.csv')  # Reemplazar con la ruta real del dataset

# üéØ Definir X y y
X = df.drop(columns='var_y')  # üëà CAMBIAR: 'var_y' por la columna target real
y = df['var_y']               # üëà CAMBIAR: 'var_y' por la columna target real
```

```python
# üîß BLOQUE LOW CODE: Ajusta solo lo necesario (¬°no modifiques todo!)

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# üëà CAMBIA AQU√ç: Nombre de la variable target
X = df.drop(columns='var_y')   # üëà CAMBIA: 'var_y' por el nombre real de tu variable objetivo
y = df['var_y']                # üëà CAMBIA: 'var_y' por el nombre real de tu variable objetivo

# üëà CAMBIA AQU√ç: Espacio de b√∫squeda
param_grid = {
    'mlpclassifier__hidden_layer_sizes': [(50,), (100,), (100, 50)],
    'mlpclassifier__activation': ['relu', 'tanh'],
    'mlpclassifier__alpha': [0.0001, 0.001],
    'mlpclassifier__learning_rate_init': [0.001, 0.01]
}

# üîí NO MODIFIQUES DE AQU√ç PARA ABAJO

pipe = make_pipeline(
    StandardScaler(),
    MLPClassifier(max_iter=1000, random_state=42)
)

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

grid = GridSearchCV(pipe, param_grid=param_grid, scoring='accuracy', cv=cv, n_jobs=-1, verbose=2)
grid.fit(X, y)

print("üìå Mejores hiperpar√°metros encontrados:")
print(grid.best_params_)
print(f"‚úÖ Accuracy promedio (CV): {grid.best_score_:.3f}")
```

```python
# üîç Pipeline y b√∫squeda de hiperpar√°metros
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Definir pipeline
pipe = make_pipeline(
    StandardScaler(),
    MLPClassifier(max_iter=1000, random_state=42)
)

# Espacio de b√∫squeda
param_grid = {
    'mlpclassifier__hidden_layer_sizes': [(50,), (100,), (100, 50)],
    'mlpclassifier__activation': ['relu', 'tanh'],
    'mlpclassifier__alpha': [0.0001, 0.001],
    'mlpclassifier__learning_rate_init': [0.001, 0.01]
}

# Cross-validation 10 folds
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# B√∫squeda
grid = GridSearchCV(pipe, param_grid=param_grid, scoring='accuracy', cv=cv, n_jobs=-1, verbose=2)
grid.fit(X, y)

# Resultados
print("üìå Mejores hiperpar√°metros encontrados:")
print(grid.best_params_)
print(f"‚úÖ Accuracy promedio (CV): {grid.best_score_:.3f}")
```
